```python

from collections import Counter
import random
import pandas as pd
import numpy as np
```

```python
N =  10000

result = []

for iter in range(N):
    dice_1 = random.randint(1,6)
    dice_2 = random.randint(1,6)

    result.append(dice_1 + dice_2)

stat = Counter(result)

df = pd.DataFrame.from_dict(stat, orient='index', columns=['count']).reset_index()
df = df.rename(columns={'index': 'sum'})
df['Probability'] = [ x / N * 100 for x in df['count']]
df['Probability_txt'] = [ f'{x} / {N}' for x in df['count']]
df = df.sort_values(by=['sum'])
df
```

```python
from matplotlib import pyplot as plt
df.plot(kind='scatter', x='sum', y='Probability', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```
